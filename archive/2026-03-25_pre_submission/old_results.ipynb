## Part 1: Empirical Results Summary

### 1. Primary Identification: The 2021 "Knowledge Wall"
We test for a discontinuity in adoption at the September 2021 LLM training cutoff.

| Estimator | Horizon (Outcome) | Sample | Estimate (Log) | P-value | Verdict |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **rdrobust (Primary)** | 52 Weeks | Full | **+0.035** | **0.589** | **Null** |
| **Clustered WLS** | 52 Weeks | Full | **+0.080** | **0.663** | **Null** |
| **rdrobust** | 52 Weeks | Min 100 DL | **+0.160** | **0.003** | **Positive** (Freshness Premium) |

**Summary:** No robust evidence of a "Knowledge Wall" in the broad population. For successful libraries, we observe a sensitive "Freshness Premium."

### 2. Robustness: Permutation Inference
We ran 366 random "Monday Cutoffs" across 7 years to see if the September 2021 result is unusual.
*   **Result:** 60% of random weeks had a larger adoption drop than the 2021 cutoff.
*   **Verdict:** The 2021 discontinuity is statistically indistinguishable from background noise (**$p = 0.60$**).

## Part 2: Technical Rationale & Defensibility

### 1. Distinction: Horizons vs. Bandwidths (CRITICAL)
*   **Outcome Horizons (12, 26, 52 weeks):** This is **what we measure**. It is the length of time we track a library *after* it is released. We use all three to study **Catch-up Dynamics** (i.e., whether the cutoff effect is temporary or permanent).
*   **Estimation Bandwidth ($\pm 26$ weeks):** This is **who we include**. It is the "window" of the running variable around the cutoff. We use $\pm 26$ weeks as our **Primary Specification** to balance statistical power with proximity to the cutoff.

### 2. The "First Observed Download" Proxy
*   **Why?** PyPI metadata (e.g., "created_at") is noisy; names can be reserved months before code is pushed.
*   **Defense:** Using the first week with positive downloads is a **behavioral measure** of when a library actually entered the ecosystem and became available for adoption.

### 3. The "Donut" RDD Rationale ($\pm 2$ weeks)
*   **Why?** LLM cutoffs are not "sharp" in reality (varying model architectures, filtering, and "cut" dates).
*   **Defense:** Excluding the boundary (±2 weeks) ensures we compare "definitely treated" (pre-cutoff) vs. "definitely control" (post-cutoff) libraries, removing the ambiguity of libraries released *during* the model's training window.

### 4. The "Zombie" Filter (`min100` downloads)
*   **Why?** Why highlight results for libraries with $>100$ downloads?
*   **Defense:** 95% of PyPI packages are "dead on arrival." If the LLM effect is a **discovery tax**, it can only be detected among libraries that are actually being discovered. We must show the findings hold for the **economically relevant** part of the ecosystem.

### 5. Clustered Inference (Weekly Clusters)
*   **Why?** Why cluster by week of release?
*   **Defense:** Libraries released in the same week share ecosystem shocks (e.g., Python version launches). Clustering by the running variable (weeks) is the **standard of care** to avoid underestimating the noise and finding "false significance."

## Part 3: Final Synthesis
1.  **Resilience:** The broad software ecosystem is remarkably resilient to LLM knowledge gaps.
2.  **Relative Suppression:** The "cost" of the cutoff is likely a **dampening of growth momentum** rather than absolute exclusion.
3.  **Freshness Premium:** Among successful tools, "newness" is a stronger signal than "LLM-knowability."
4.  **Catch-up:** Any exclusion costs appear to be temporary, recovering within the first year of diffusion.
